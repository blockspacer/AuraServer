```python
import math
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
ROW_HEIGHT = 100
SIDE = ROW_HEIGHT / math.sqrt(3)
DIAGONAL = 2 * SIDE

pointy = [math.sqrt(3.0), math.sqrt(3.0) / 2.0, 0.0, 3.0 / 2.0, math.sqrt(3.0) / 3.0, -1.0 / 3.0, 0.0, 2.0 / 3.0, 0.5]
flat = [3.0 / 2.0, 0.0, math.sqrt(3.0) / 2.0, math.sqrt(3.0), 2.0 / 3.0, 0.0, -1.0 / 3.0, math.sqrt(3.0) / 3.0, 0.0]

def p2h(x, y, o=flat):
    pt = [(x - 0) / 15.0, (y - 0) / 15.0]
    q = o[4] * pt[0] + o[5] * pt[1]
    r = o[6] * pt[0] + o[7] * pt[1]
    s = -q -r
    
    # Round
    q_int = int(round(q))
    r_int = int(round(r))
    s_int = int(round(s))
    
    q_diff = abs(q_int - q)
    r_diff = abs(r_int - r)
    s_diff = abs(s_int - s)
    
    if q_diff > r_diff and q_diff > s_diff:
        q_int = -r_int - s_int
    else:
        if r_diff > s_diff:
            r_int = -q_int - s_int
        else:
            s_int = -q_int - r_int
            
    return q_int, r_int, s_int
```

```python
p2h(1.0, 38.8)
```

```python
start = -20
end = 20
total = (end - start)

x = np.linspace(start, end, total)
y = np.linspace(start, end, total)

grid = [[p2h(xi, yi) for xi in x] for yi in y]
grid = np.asarray(grid)
```

```python
grid_f = grid.astype(np.float)
grid_f = grid_f / grid_f.max()

fig = plt.figure()
fig.set_size_inches((16,16))
plt.imshow(grid_f, interpolation='nearest')
```
